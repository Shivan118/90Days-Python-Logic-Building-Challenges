# Question

- Finding Missing Number in a List step wise apparch 

### <span style="color: red;"> Steps to Solve the Problem </span>


**1. Understand the Input and Output**

- **Input:** A list nums containing n-1 integers ranging from 1 to n.
- **Output:** The missing integer.

**2. Calculate the Expected Sum**

The sum of the first `n` natural numbers can be calculated using the formula: Sum expected = ð‘› Ã— ( ð‘› + 1 ) / 2 

**3. Calculate the Actual Sum**

Sum the elements in the given list nums to get the `Sum_actual`

**4. Find the Missing Number**

The missing number can be determined by subtracting the `Sum_actual` from `Sum_expected`:

MissingÂ Number = `Sum expected` âˆ’ `Sum actual`

```python
from typing import List

def find_missing_number(nums: List[int], n: int) -> int:
    # Step 2: Calculate the expected sum of the first n natural numbers
    sum_expected = n * (n + 1) // 2
    
    # Step 3: Calculate the actual sum of the numbers in the list
    sum_actual = sum(nums)
    
    # Step 4: The missing number is the difference between expected sum and actual sum
    missing_number = sum_expected - sum_actual
    
    return missing_number
```

```python
# Example input
nums = [1, 2, 4, 5, 6]
n = 6

# Call the function
print(find_missing_number(nums, n))  # Output: 3
```
